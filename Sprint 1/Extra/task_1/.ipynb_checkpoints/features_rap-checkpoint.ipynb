# Setup

```python
import ee
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance

ee.Authenticate()
ee.Initialize(project="shrubwise-dc-488219")

RAP_veg_yearly_10m = ee.ImageCollection("projects/rap-data-365417/assets/vegetation-cover-10m")
```

# Single Case

```python
# D.L. Bliss AOI bbox
aoi = ee.Geometry.Rectangle([-120.1018846, 38.99274873, -120.0899834, 39.0020357], geodesic=False)
```

```python
def rap_shrub_for_year_aoi(year, aoi):
    
    ic = (RAP_veg_yearly_10m
          .filter(ee.Filter.eq("year", year))
          .filterBounds(aoi))

    # Guard: if no tile intersects, fail loudly (better than silent None stats)
    size = ic.size()
    if size.getInfo() == 0:
        raise ValueError(f"No RAP 10m tiles found for year={year} intersecting AOI.")

    def add_intersection_area(img):
        inter = img.geometry().intersection(aoi, ee.ErrorMargin(1))
        return img.set("aoi_intersection_area_m2", inter.area(1))

    tile = ee.Image(ic.map(add_intersection_area).sort("aoi_intersection_area_m2", False).first())
    return tile.select("SHR").rename("shrub_cover").set("year", year)

shrub_2025 = rap_shrub_for_year_aoi(2025, aoi)
print("Band names:", shrub_2025.bandNames().getInfo())
print("Valid pixel count in AOI:",
      shrub_2025.reduceRegion(ee.Reducer.count(), aoi, 10, maxPixels=1e9, bestEffort=True).get("shrub_cover").getInfo())
```

```python
# Multi-year stack: 2018â€“2026
years = list(range(2018, 2026))
imgs = [rap_shrub_for_year_aoi(y, aoi) for y in years]
stack = ee.ImageCollection(imgs)

# Mean / std
mean_img = stack.mean().rename("rap_shrub_mean_2018_2025")
std_img  = stack.reduce(ee.Reducer.stdDev()).rename("rap_shrub_std_2018_2025")

# Trend: linear fit (slope per year)
def add_t_float(im):
    t = ee.Image.constant(ee.Number(im.get("year"))).toFloat().rename("t")
    y = im.select("shrub_cover").toFloat()
    return y.addBands(t)

with_t = stack.map(add_t_float)

fit = with_t.select(["t", "shrub_cover"]).reduce(ee.Reducer.linearFit())
trend_img = fit.select("scale").rename("rap_shrub_trend_per_year_2018_2025")

# Change: 2025 - 2018
change_img = (
    rap_shrub_for_year_aoi(2025, aoi)
      .subtract(rap_shrub_for_year_aoi(2018, aoi))
      .rename("rap_shrub_change_2025_minus_2018")
)

# Threshold masks
mask_ge5  = rap_shrub_for_year_aoi(2025, aoi).gte(5).rename("rap_shrub_ge5_2025")
mask_ge10 = rap_shrub_for_year_aoi(2025, aoi).gte(10).rename("rap_shrub_ge10_2025")

feature_cube = ee.Image.cat([mean_img, std_img, change_img, trend_img, mask_ge5, mask_ge10])
print('Feature cube bands:', feature_cube.bandNames().getInfo())
```

```python
reducer = (ee.Reducer.mean()
           .combine(ee.Reducer.median(), sharedInputs=True)
           .combine(ee.Reducer.min(), sharedInputs=True)
           .combine(ee.Reducer.max(), sharedInputs=True))

summary = feature_cube.reduceRegion(
    reducer=reducer,
    geometry=aoi,
    scale=10,
    maxPixels=1e9,
    bestEffort=True
).getInfo()

for k in sorted(summary.keys()):
    print(k, summary[k])
```

# Generic dist comparison

```python
# AOI 1: D.L. Bliss (given)
aoi_bliss = ee.Geometry.Rectangle(
    [-120.1018846, 38.99274873, -120.0899834, 39.0020357],
    geodesic=False
)

# AOI 2: "Shifted" AOI about ~10 km east
minx, miny, maxx, maxy = -120.1018846, 38.99274873, -120.0899834, 39.0020357
center_lat = (miny + maxy) / 2
km_shift = 10
dlon = km_shift / (111.32 * np.cos(np.deg2rad(center_lat)))

aoi_demo_shifted = ee.Geometry.Rectangle([minx + dlon, miny, maxx + dlon, maxy], geodesic=False)

AOIS_DEFAULT = {
    "DL_Bliss": aoi_bliss,
    "Demo_Shifted_10km": aoi_demo_shifted
}
```

```python
def rap_shrub_for_year_mosaic(year):
    ic = RAP_veg_yearly_10m.filter(ee.Filter.eq("year", year))
    return ic.mosaic().select("SHR").rename("shrub_cover").set("year", year)

years = list(range(2018, 2026))
stack = ee.ImageCollection([rap_shrub_for_year_mosaic(y) for y in years])

mean_img = stack.mean().rename("rap_shrub_mean_2018_2025")
std_img  = stack.reduce(ee.Reducer.stdDev()).rename("rap_shrub_std_2018_2025")
change_img = rap_shrub_for_year_mosaic(2025).subtract(rap_shrub_for_year_mosaic(2018)).rename("rap_shrub_change_2025_minus_2018")

def add_t_float(im):
    t = ee.Image.constant(ee.Number(im.get("year"))).toFloat().rename("t")
    y = im.select("shrub_cover").toFloat()
    return y.addBands(t)

fit = stack.map(add_t_float).select(["t", "shrub_cover"]).reduce(ee.Reducer.linearFit())
trend_img = fit.select("scale").rename("rap_shrub_trend_per_year_2018_2025")

mask_ge5  = rap_shrub_for_year_mosaic(2025).gte(5).rename("rap_shrub_ge5_2025")
mask_ge10 = rap_shrub_for_year_mosaic(2025).gte(10).rename("rap_shrub_ge10_2025")

feature_cube = ee.Image.cat([mean_img, std_img, change_img, trend_img, mask_ge5, mask_ge10])
```

```python
def _wasserstein_1d(a, b):
    return float(wasserstein_distance(a, b))

def _psi(a, b, bins=20, eps=1e-6):
    # robust bins via combined quantiles (handles skew + many zeros)
    qs = np.linspace(0, 1, bins + 1)
    edges = np.quantile(np.concatenate([a, b]), qs)
    edges = np.unique(edges)
    if edges.size < 3:
        return 0.0

    ha, _ = np.histogram(a, bins=edges)
    hb, _ = np.histogram(b, bins=edges)

    pa = ha / max(ha.sum(), 1)
    pb = hb / max(hb.sum(), 1)

    pa = np.clip(pa, eps, None)
    pb = np.clip(pb, eps, None)

    return float(np.sum((pa - pb) * np.log(pa / pb)))


def compare_feature_distributions(
    image,
    band,
    aois,
    scale = 10,
    n_samples = 5000,
    seed = 42,
    psi_bins = 20,
    hist_bins = 50,
    hist_range = (0, 100),
    make_plots = True,
):
    band_img = image.select(band).toFloat()

    def sample_values(region: ee.Geometry) -> np.ndarray:
        pts = ee.FeatureCollection.randomPoints(region=region, points=n_samples, seed=seed)
        fc = band_img.sampleRegions(collection=pts, scale=scale, geometries=False)
        vals = fc.aggregate_array(band).getInfo()
        arr = np.array(vals, dtype=float)
        arr = arr[np.isfinite(arr)]
        return arr

    # sample and summarize
    samples = {}
    summary_rows = []
    for name, geom in aois.items():
        arr = sample_values(geom)
        samples[name] = arr

        pct_gt0 = float(np.mean(arr > 0)) if arr.size else np.nan

        summary_rows.append({
            "aoi": name,
            "n": int(arr.size),
            "mean": float(np.mean(arr)) if arr.size else np.nan,
            "median": float(np.median(arr)) if arr.size else np.nan,
            "p05": float(np.quantile(arr, 0.05)) if arr.size else np.nan,
            "p25": float(np.quantile(arr, 0.25)) if arr.size else np.nan,
            "p75": float(np.quantile(arr, 0.75)) if arr.size else np.nan,
            "p95": float(np.quantile(arr, 0.95)) if arr.size else np.nan,
            "min": float(np.min(arr)) if arr.size else np.nan,
            "max": float(np.max(arr)) if arr.size else np.nan,
            "pct_gt0": pct_gt0,
        })

    summary_df = pd.DataFrame(summary_rows).set_index("aoi")

    names = list(aois.keys())
    shift_rows = []
    for i in range(len(names)):
        for j in range(i + 1, len(names)):
            a_name, b_name = names[i], names[j]
            a, b = samples[a_name], samples[b_name]
            if a.size == 0 or b.size == 0:
                w, p = np.nan, np.nan
            else:
                w = _wasserstein_1d(a, b)
                p = _psi(a, b, bins=psi_bins)

            shift_rows.append({
                "aoi_a": a_name,
                "aoi_b": b_name,
                "wasserstein": w,
                "psi": p,
            })

    shift_df = pd.DataFrame(shift_rows)

    if make_plots:
        plt.figure()
        for name, arr in samples.items():
            plt.hist(arr, bins=hist_bins, range=hist_range, alpha=0.5, label=name)
        plt.title(f"Distribution comparison: {band}")
        plt.xlabel(band)
        plt.ylabel("Count")
        plt.legend()
        plt.show()

    return summary_df, shift_df, samples
```

```python
summary_df, shift_df, samples = compare_feature_distributions(
    image=feature_cube,
    band="rap_shrub_mean_2018_2025",
    aois=AOIS_DEFAULT
)

summary_df
```
