```python
fp = ca_100m_dir / "Landsat_Composite_2020_2020_26-02-19-03-59-50.tif"
fp
```

```python
import os

check_fp = os.listdir(ca_100m_dir)
```

```python
import rasterio
import matplotlib.pyplot as plt

with rasterio.open(fp) as src:
    print(f"{src.count} bands")
    band1 = src.read(1)

    plt.imshow(band1, cmap='pink')
    plt.show()
```
