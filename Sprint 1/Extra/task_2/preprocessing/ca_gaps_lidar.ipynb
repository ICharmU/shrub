```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. Load California geometry from a public GeoJSON
states_url = "https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json"
usa = gpd.read_file(states_url)
ca = usa[usa['name'] == 'California']

# 2. MATCH THE CRS (CRITICAL STEP)
# Your .laz gdf likely uses a projected CRS (meters). 
# We must reproject California to match your data.
ca = ca.to_crs(gdf.crs)

# 3. Create the plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot California as the bottom layer
ca.plot(ax=ax, color='whitesmoke', edgecolor='black', zorder=1)

# Plot your LiDAR points on top
gdf.plot(ax=ax, cmap='Blues', markersize=0.1, zorder=2)
gdf2.plot(ax=ax, cmap='Reds', markersize=0.1, zorder=2)

plt.title("LiDAR Points over California Boundary")
plt.show()
```
